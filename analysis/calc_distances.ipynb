```python
import numpy as np
import pandas as pd
from scipy.spatial import distance_matrix
from sklearn.metrics.pairwise import cosine_distances
```

```python
from pathlib import Path

Path("../data/analysis/").mkdir(parents=True, exist_ok=True)
```

```python
df_tfidf = pd.read_csv("../data/vocabulary/tfidf_matrix.csv", index_col=0)
df_tfidf
```

```python
distances_tfidf=pd.DataFrame(
    cosine_distances(df_tfidf.values, df_tfidf.values),
    index=df_tfidf.index,
    columns=df_tfidf.index
)
distances_tfidf
```

```python
distances_tfidf.to_csv("../data/analysis/distances_tfidf_cosine.csv")
```

```python
embeddings_bert = np.load("../data/embeddings/embeddings_bert.npy")
df_bert = pd.DataFrame(embeddings_bert)
df_bert
```

```python
distances_bert=pd.DataFrame(
    cosine_distances(df_bert.values, df_bert.values),
    index=df_bert.index,
    columns=df_bert.index
)
distances_bert
```

```python
distances_bert.to_csv("../data/analysis/distances_bert_cosine.csv")
```
